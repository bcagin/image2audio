```python
data_path = "/content/train_set"  # data folder
annotations_path = "/content/train_info.csv"  # data csv
```

```python
import os
import clip
import torch
from PIL import Image
import pandas as pd
from torch.utils.data import Dataset, DataLoader
import torch.nn as nn

# set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# CLIP model
clip_model, clip_preprocess = clip.load("ViT-B/32", device=device)

# CLIP float32
clip_model = clip_model.float()

# read annotation
annotations = pd.read_csv(annotations_path, header=None, names=['image_name', 'category'])

# Dataset class
class FoodDataset(Dataset):
    def __init__(self, annotations, images_path, transform=None):
        self.annotations = annotations
        self.images_path = images_path
        self.transform = transform

    def __len__(self):
        return len(self.annotations)

    def __getitem__(self, idx):
        img_name = self.annotations.iloc[idx, 0]
        img_path = os.path.join(self.images_path, img_name)  # os.path.join ile dosya yolunu birleştirme
        image = Image.open(img_path).convert("RGB")

        if self.transform:
            image = self.transform(image)

        label = int(self.annotations.iloc[idx, 1])

        return image, label

# Dataset ve DataLoader
dataset = FoodDataset(annotations=annotations, images_path=data_path, transform=clip_preprocess)
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)

# CLIP model
class CLIPClassificationModel(nn.Module):
    def __init__(self, num_classes):
        super(CLIPClassificationModel, self).__init__()

        # CLIP model Load
        self.clip_model = clip_model
        for param in self.clip_model.parameters():
            param.requires_grad = False

        # Görüntü özelliklerini sınıflandırma için tam bağlantılı katmana aktar
        self.fc = nn.Linear(self.clip_model.visual.output_dim, num_classes)

    def forward(self, images):
        # CLIP ile özellik çıkarma
        features = self.clip_model.encode_image(images).float()  # float32'e dönüştür

        # Tam bağlantılı katman ile kategori tahmini
        outputs = self.fc(features)

        return outputs

# Model, loss fonksiyonu ve optimizer tanımlama
num_classes = 251  # Toplam kategori sayısı
model = CLIPClassificationModel(num_classes=num_classes).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.fc.parameters(), lr=0.001)  # Sadece fc katmanını optimize ediyoruz
```

```python
# Doğrulama veri yollarını tanımlayın
val_data_path = "/content/val_set"  # Doğrulama veri klasörünün yolu
val_annotations_path = "/content/val_info.csv"  # Doğrulama CSV dosyasının yolu

# Doğrulama annotation dosyasını okuyun
val_annotations = pd.read_csv(val_annotations_path, header=None, names=['image_name', 'category'])

# Doğrulama için Dataset ve DataLoader
val_dataset = FoodDataset(annotations=val_annotations, images_path=val_data_path, transform=clip_preprocess)
val_dataloader = DataLoader(val_dataset, batch_size=32, shuffle=False)
```

```python
# Eğitim döngüsü
num_epochs = 10

for epoch in range(num_epochs):
    # Eğitim aşaması
    model.train()
    total_loss = 0.0
    for images, labels in dataloader:
        images = images.to(device).to(torch.float32)
        labels = labels.to(device)

        # İleri yönlü geçiş
        outputs = model(images)
        loss = criterion(outputs, labels)

        # Geriye yayılım ve optimizasyon
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    avg_train_loss = total_loss / len(dataloader)
    print(f"Epoch [{epoch+1}/{num_epochs}], Training Loss: {avg_train_loss:.4f}")

    # Doğrulama aşaması
    model.eval()  # Modeli değerlendirme moduna geçir
    val_loss = 0.0
    correct = 0
    total = 0
    with torch.no_grad():  # Doğrulama sırasında gradyan hesaplanmaz
        for images, labels in val_dataloader:
            images = images.to(device).to(torch.float32)
            labels = labels.to(device)

            # Doğrulama veri setinde tahminleri yap
            outputs = model(images)
            loss = criterion(outputs, labels)
            val_loss += loss.item()

            # Doğruluğu hesaplamak için tahmin edilen sınıfları alın
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    # Ortalama doğrulama kaybı ve doğruluk
    avg_val_loss = val_loss / len(val_dataloader)
    val_accuracy = 100 * correct / total
    print(f"Validation Loss: {avg_val_loss:.4f}, Validation Accuracy: {val_accuracy:.2f}%")
```
